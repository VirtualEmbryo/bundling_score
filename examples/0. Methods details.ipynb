# Introduction

Fibers orientation was extracted from the image intensity $I(r)$ and its gradient $∇I$. The local direction of the gradient was defined as

$$n=\frac{\nabla I}{∥\nabla I∥}$$

The local orientation of the fibers $u(r)$ is expected to be orthogonal to $n$. Based on this, the nematic tensor Q:

$$Q_{ij}(r)=  \frac{δ_{ij}}{2}-n_i n_j$$	

The spatial correlations of $Q$ were analyzed using the correlation coefficient C, which ranges from -1 to 1 and measures the alignment of orientations with a shifted version of the image. A value of 1 for C means that orientations are completely correlated, while a value of -1 indicates that orientations are anti-correlated, and a value of 0 means that there is no correlation.

Correlations are computed on a square window of half-length $s_{max}<a$, where $a$ is the half-length of the image.

$$C(x,y)=\frac{\iint_{-(a-s_{max})}^{a-s_{max}} Q_{ij} (x_0, y_0 )Q_{ij} (x_0+x, y_0+y)dx_0dy_0}{\iint Q_{ij}^2 (x_0,y_0)dx_0dy_0}$$

The mean area over which the orientation is correlated is obtained by integrating C over the window:

$$A=\iint_{-s_{max}}^{s_{max}}C(x,y) dxdy$$
